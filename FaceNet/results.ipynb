# Loading and showing results

```python
import csv

RESULTS_FILE = './results/test_uncloaked_high.csv' # csv file predictions, true value, distance

predictions =[]
truth = []
distance = []

with open(RESULTS_FILE) as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        predictions.append(row[0])
        truth.append(row[1])
        distance.append(row[2])
```

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, classification_report # evaluate the results using metrics precsion, recall, f1-score
import matplotlib.pyplot as plt

print("Score: ", accuracy_score(truth, predictions))
print(classification_report(truth, predictions))
```

```python
disp = ConfusionMatrixDisplay.from_predictions(truth, predictions, cmap=plt.cm.Blues) # draw a confusion matrix
disp.ax_.set_title('Facenet trained from Scratch')
disp.figure_.set_size_inches((15, 15))
plt.show()
```
